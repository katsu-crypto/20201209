```python
%reset
#
import numpy as np
N_train=800#
N_test=200#

np.random.seed(1)
x_train=(np.random.rand(N_train,2)*1000).astype(np.int32)+1
y_train=(x_train.min(axis=1)/x_train.max(axis=1)<=0.2).astype(np.int)

x_test=(np.random.rand(N_test,2)*1000).astype(np.int32)+1
y_test=(x_test.min(axis=1)/x_test.max(axis=1)<=0.2).astype(np.int)

from tensorflow.python.keras.utils import to_categorical
y_train=to_categorical(y_train,2)
y_test=to_categorical(y_test,2)


#---------------------------------------------
from tensorflow.python.keras.models import Sequential
model=Sequential()


#
from tensorflow.python.keras.layers import Dense

model.add(
    Dense(
        units=2,
        input_shape=(2,),
        activation="sigmoid"
    )
)
model.add(
    Dense(
        units=2,
        activation='sigmoid'
    )
)
#
model.add(
    Dense(
        units=2,
        #activation='sigmoid'
        activation='softmax'#softmax
    )
)

#---------------------------------_--0----------------
from tensorflow.python.keras.callbacks import TensorBoard

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

tsb=TensorBoard(log_dir='./logs')
history_adam=model.fit(
    x_train,
    y_train,
    batch_size=100,
    epochs=3000,
    validation_split=0.2,
    callbacks=[tsb]
)
```

```python
expect=model.predict(x_test)
```

```python
result=model.evaluate(x_test,y_test)
```
